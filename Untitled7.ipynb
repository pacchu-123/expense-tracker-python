```python
import pandas as pd

data = {
    "Date": ["2026-03-01","2026-03-02","2026-03-02","2026-03-03","2026-03-04"],
    "Category": ["Food","Travel","Food","Shopping","Food"],
    "Amount": [200,150,300,500,250]
}

df = pd.DataFrame(data)
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])

print("Cleaned Data")
df
```

```python
total_expense = df["Amount"].sum()

print("Total Expense:", total_expense)
```

```python
category_expense = df.groupby("Category")["Amount"].sum()

print(category_expense)
```

```python
import matplotlib.pyplot as plt

category_expense.plot(kind="pie", autopct='%1.1f%%')

plt.title("Expense Distribution")
plt.ylabel("")
plt.show()
```

```python
category_expense.plot(kind="bar")

plt.title("Expense by Category")
plt.xlabel("Category")
plt.ylabel("Amount")

plt.show()
```

```python
budget = 1000

if total_expense > budget:
    print("⚠ Budget exceeded!")
else:
    print("✅ Budget within limit")
```

```python
df.to_excel("expense_report.xlsx", index=False)

print("Excel report saved")
```
